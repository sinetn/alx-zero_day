## Lex only

```python
#@title Writing Lex program
%%writefile program.l

%{
    #include <stdio.h>
    int ctChar=0;
    int ctSpace=0;
    int ctWord=0;
    int ctLine=0;
%}
WORD [^ \t\n,\.:]+
EOL [\n]
BLANK [ ]
%%

{WORD} {ctWord++; ctChar+=yyleng;}
{BLANK} {ctSpace++;}
{EOL} {ctLine++;}
. {ctChar++;}
%%

void main(int argc, char *argv[]){
    if(argc!=2){
        printf("Usage:\n\t./a.out <FILENAME>\n");
        exit(0);
    }

    yyin=fopen(argv[1],"r");
    yylex();

    printf("Word Count: %d\n",ctWord);
    printf("Character Count: %d\n",ctChar);
    printf("Space Count: %d\n",ctSpace);
    printf("Line Count: %d\n",ctLine);
    fclose(yyin);

}

int yywrap(){
    return 1;
}
```

```python
#@title Shell Execution (you can rewrite the commands as per your need, eg. if you want to include a file as an input)
%%shell

lex -l program.l
gcc lex.yy.c
./a.out program.txt
```

## Lex and Yacc combined

```python
#@title Writing YACC program
%%writefile program.y

%{
    #include<stdio.h>
    #include<stdlib.h>
    int yyerror();
    extern int yylex(); //scanner
    extern int line; // imported from the scanner
%}
%token START END ID NUM PLUS MINUS MULTIPLY DIVIDE ASSIGN

%%
Program: START Statements END
Statements: Statement | Statement Statements

Statement: ID ASSIGN Expression
Expression: Expression PLUS Term | Expression MINUS Term | Term
Term: Term MULTIPLY Factor | Term DIVIDE Factor | Factor
Factor: ID | NUM

%%

int yyerror(){
    printf("Invalid Identifier\n");
    exit(0);
}

void main(){
    printf("Compiler starts Lexical and Syntax analysis ... \n");
    yyparse();
    printf("Compiler completed Lexical and Syntax analysis ... \n");

}
```

```python
#@title Writing Lex program
%%writefile program.l

%{
    #include "y.tab.h"
    int line = 1;
%}
%%
Start {printf{"Scanner: recognizing Token START\n"}; return START;}
End {printf{"Scanner: recognizing Token END\n"}; return END;}
[a-zA-Z][a-zA-Z0-9] {printf{"Scanner: recognizing Token ID\n"}; return ID;}
[0-9]* {printf{"Scanner: recognizing Token NUM\n"}; return NUM;}
"+" {printf{"Scanner: recognizing Token PLUS\n"}; return PLUS;}
"-" {printf{"Scanner: recognizing Token MINUS\n"}; return MINUS;}
"*" {printf{"Scanner: recognizing Token MULTIPLY\n"}; return MULTIPLY;}
"\" {printf{"Scanner: recognizing Token DIVIDE\n"}; return DIVIDE;}
"=" {printf{"Scanner: recognizing Token ASSIGN\n"}; return ASSIGN;}
[\n] {line++;}
[ \t] {;}
. {printf{"Lexical Error (Line %d): %s", line, yytext};}
%%
```
